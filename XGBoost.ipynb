# XGBRegressor Model

```python
#imported python packages
import numpy as np
import pandas as pd
from math import sqrt

from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from xgboost import plot_importance
from sklearn.metrics import roc_curve, auc

import time
import pickle
import joblib
```

```python
# import dataset about train dataset 
train_df = pd.read_csv("train_df.csv")
```

```python
train_df.info()
train_df
```

```python
# import dataset about test dataset
test_df = pd.read_csv("test_df.csv")
```

```python
test_df.info()
test_df
```

```python
# Splitting the train dataset into test, validation, and train dataset
X_train, X_test, Y_train, Y_test = train_test_split(train_df.drop(columns=["segment_id","time_to_eruption"]),
                                                    train_df[["time_to_eruption"]],
                                                    test_size=0.20)
X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size=0.20)
```

```python
# instantiated XGB model
model = XGBRegressor(
    learning_rate=0.05,
    n_estimators=200,
    seed=42,
    eval_metric="mae",
    max_depth=5,
    eta=0.2,
    gamma=4,
    min_child_weight=6,
    subsample=0.8,
    #objective='reg:linear',
    early_stopping_rounds=10,
    num_round=200)
```

```python
# train the model
model.fit(
    X_train, 
    Y_train, 
    eval_set=[(X_train, Y_train), (X_val, Y_val)])
```

```python
# get the predictions for the validation data
Y_pred = model.predict(X_val)
```

```python
Y_pred
```

```python
# MAE metric for the predicted and actual data
print('XGBoost MAE =', mean_absolute_error(Y_val,Y_pred))
```

```python
# importance of each feature plotted
fig, ax = plt.subplots(1,1,figsize=(10,14))
plot_importance(booster=model, ax=ax)
```
